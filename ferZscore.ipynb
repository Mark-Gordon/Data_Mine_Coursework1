```python
import pandas as pd
from scipy.stats import zscore
```

```python
df = pd.read_csv("fer2018PixelSplitRandomize.csv");
```

```python
df.head()
```

```python
index_df = df["emotion"]
df = df.drop(["emotion"], axis=1)
```

```python
df = df.apply(zscore)
```

```python
df.insert(0, "emotion", index_df)
df = df.set_index("emotion")
```

```python
df.to_csv("fer2018PixelSplitRandomizeZscore.csv");
```
